```python
from matplotlib import pyplot as plt
from PIL import Image
from numpy import asarray
from numpy import array
from keras.models import load_model
from numpy import expand_dims
from numpy import reshape
from numpy import load
from numpy import max
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import Normalizer
from sklearn.svm import SVC
import cv2
from facenet_pytorch.models.mtcnn import MTCNN
import numpy as np
import torch
import math
```

```python
device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
mtcnn = MTCNN(keep_all=True, device=device)
def extract_image(image):
  img1 = Image.open(image)
  img1 = img1.convert('RGB')
  pixels = asarray(img1)
  boxes, probs, landmarks = mtcnn.detect(pixels, landmarks=True)
  face_images = []
  for box, prob, ld in zip(boxes, probs, landmarks):
    x1 = int(abs(math.ceil(box[0])))
    y1 = int(abs(math.ceil(box[1])))
    x2 = int(abs(math.ceil(box[2])))
    y2 = int(abs(math.ceil(box[3])))
    # print(x1,y1,x2,y2)
    store_face = pixels[y1:y2,x1:x2]
    if len(store_face) == 0:
      continue
    image1 = Image.fromarray(store_face,'RGB')
    image1 = image1.resize((160,160))
    face_array = asarray(image1)
    face_images.append(face_array)
  return face_images
```

```python
from os import listdir
def load_faces(directory):
  face = []
  i=1
  for filename in listdir(directory):
    path = directory + filename
    faces = extract_image(path)
    face.extend(faces)
  return face
```
